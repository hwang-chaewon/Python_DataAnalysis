# **들여쓰기**

- **공백 문자**(스페이스나 탭)를 사용하여 코드를 그룹화하여 구조화함
  - 콜론(**:**)은 코드 블록의 시작을 표현 
  - 같은 블록에 있는 코드들은 모두 같은 크기만큼 들여쓰기해야 함
  - 가독성이 증가함 
 

*   자바 같은 경우, 코드가 끝날 때마다 세미콜론(;)을 뒤에 넣어 컴파일러(컴퓨터가 실행할 수 있는 코드로 바꿔주는 것)가 인식하게 함
*   파이썬은 들여쓰기를 사용해 문장/코드가 끝났다는 것 표시. 파이썬 같은 경우 들여쓰기가 필수이므로 가독성이 증가.
*   다른 코딩 언어에선 들여쓰기는 개발자가 가독성을 위해 자의로 하는 것이지 안 할 경우 문제가 생기지 않음.



````python
less = []
greater = []

#원래 이 선 없었는데 도구>설정에서 들여쓰기 가이드를 설정했더니 생김. 파데분 수업 좋다. 많은 거 알아가네...
for x in range(10):
    if x < 5:
        less.append(x)
    else:
        greater.append(x)

print(less)
print(greater)
````

- 세미콜론(**;**)
  - 파이썬은 한 줄에 한 문장을 작성
  - 한 줄에 여러 문장을 작성하려면 각 문장을 구별하기 위해 세미콜론을 사용함

````python
a=1; b=2; c=3
````

# **튜플(Tuple)**

## **튜플**

- 리스트와 같은 데이터의 목록
 - 리스트와 거의 동일하다. 다른 점은 리스트는 mutable [ ], 튜플은 immutable ()
 - 콤마(,)로 구별하여 데이터를 나열하여 표시
 - 괄호()안에 콤마로 구별하여 나열하여 표시
 - 길이 1개짜리 tuple은 반드시 뒤에 콤마(,)를 표시. 
   -  예를 들어 (1)이라면, (1)은 튜플인지 그냥 숫자 1인지 구분할 수 없어 튜플은 (1,)로 표기해야 한다
 - 리스트와 같은 방법으로 원소에 접근
 - 리스트와 다르게 immutable 객체: 한 번 만든 튜플은 내용을 수정하지 못함
- 튜플 유용성
 - 연관된 정보를 묶어서 저장함과 동시에 정보 수정의 위험을 제거함

````python
t = ( 2022, '학기', 1 , '분반' )
tinyt = ('기초', '파이썬과데이터분석')
````

````python
# 전체 튜플 출력
print (t)   

# 인덱싱을 통해 튜플의 첫 번째 원소 접근    
print (t[0])    

# 슬라이싱을 통해 튜플의 두 번째 원소부터 세 번째 원소 전까지 접근  
print (t[1:3])

# 슬라이싱을 통해 튜플의 두 번째 원소부터 세 번째 원소 전까지 접근  
print (t[2:])     

# * 튜플의 반복
print (tinyt * 2) 

# + 튜플의 결합
print (t + tinyt) 
print(t)
````

## **range**

- 연속적인 정수를 만드는 함수
- 형식
  -  range(stop): 
    - 0부터 시작하여 stop은 포함하지 않음
    - step: default로 1
  -  range(start, stop)
    - start부터 시작하여 stop은 포함하지 않음
    - step: default로 1
  -  range(start, stop, step)
    - start부터 시작하여 stop은 포함하지 않고 step은 간격 

````python
r = range(1,10)
r
````

````python
list(r)
````

````python
range(1,10,2)
````

````python
list(range(1,10,2))
````

````python
list(range(10,2,-2))
````

````python
range(10, 1, 2)
````

````python
list(range(10, 1, 2))
# 10부터 2씩 증가해서 1에 다가가는 정수는 없음
````

````python
range(1,10,-2)
````

````python
list(range(1,10,-2)) 
# 1부터 -2씩 감소해서 10에 다가가는 정수는 없음
````

### **형변환**

- list(X): X로 채워진 리스트를 만듬

- tuple(X): X로 채워진 튜플을 만듬

````python
#tuple-->list
list((1,2,3))
````

````python
list("String")
````

````python
#list-->tuple
tuple([1,2,3])
````

````python
tuple("String")
````
