## **[IF...ELIF...ELSE 문]**

순차적으로 걸러짐.

```

    statement

    statement
```

````python
amount=int(input("Enter amount: "))
if amount<1000:
    discount=amount*0.05
    print ("Discount",discount)
elif amount<5000:
    discount=amount*0.10
    print ("Discount",discount)
else:
    discount=amount*0.15
    print ("Discount",discount)
print ("Net payable:",amount-discount)
````

## **[중첩된 ELSE 문]**

````python
#Nested IF Statements 
num=int(input("enter number"))
if num%2==0:
    if num%3==0:
        print ("Divisible by 3 and 2")
    else:
        print ("divisible by 2 not divisible by 3")
else:
    if num%3==0:
        print ("divisible by 3 not divisible by 2")
    else:
        print ("not Divisible by 2 not divisible by 3")
````

## **[BREAK 문]**

- 반복문을 빠져나감
  - 중첩된 반복문일 경우 break 문을 감싼 가장 가까운 반복문을 벗어남

[FOR...BREAK  문의 예제]

````python
for letter in 'Python':     
    if letter == 'h':
        break
    print ('Current Letter :', letter)
````

[WHILE...BREAK문의 예]

````python
var = 10            # Second Example
while var > 0:
    print ('Current variable value :', var)
    var = var -1
    if var == 5:
        break
````

[FOR...ELSE...BREAK 문의 예]

````python
no=int(input('any number: '))
numbers=[11,33,55,39,55,75,37,21,23,41,13]
for num in numbers:
    if num==no:
        print ('number found in list')
        break
else:
    print ('number not found in list')
````

## **[CONTINUE 문]**


  -  continue문 다음의 실행 문장들을 건너 뛰고, 다음
        반복을 시작. 
    
    continue가 실행될 경우 그때의 반복만 안하겠다는 것


[FOR...CONTINUE  문의 예제]

````python
for letter in 'Python':     # First Example
    if letter == 'h':
        continue
    print ('Current Letter :', letter)
````

[WHILE...CONTINUE 문의 예제]

````python
var = 10            
while var > 0:
    var = var -1
    if var == 5:
        continue
    print ('Current variable value :', var)
````

# **함수**

**함수**: 입력을 받아서 어떤 작업(계산, 기능)을 수행하고 결과값을 넘겨줌

함수라기보단 그냥 '기능'이라고 생각하기

function,method, procedure라고도 함

**함수의 실행 순서**
1. 입력을 받음
2. 작업을 수행
3. 결과값을 넘겨줌

**함수의 정의**:


```
def 함수이름(매개변수):
    <수행할 문장>
    ...
    return 결과값
```


**함수의 호출**:


```
함수이름(인수)
```



- 매개변수: 함수에 입력으로 전달된 값을 받는 변수


```
def func( par1, par2):
    ...
```


- 인수: 함수를 호출할 때 전달하는 입력값


```
func(10, 20)
```

## **[전역변수와 지역변수]**

- 변수의 선언 위치에 따라
 - 지역 변수: 함수 안에 선언되는 변수
   -  함수가 호출되었을 때 만들어지고 종료하면 사라짐
   -  매개변수도 지역변수에 속함
 - 전역 변수: 함수 밖에 선언되는 변수

````python
def func(n):        
    tmp = n +1      
    print(tmp)      
func(7)
````

````python
tmp
````

````python
num = 7         
def func():
    print(num)      # 함수 내에서 전역변수에 접근 가능
func()
````

````python
num = 7           
def func():
    num = 0         # 지역 변수
    print(num)      # 지역변수 num 출력
func()
````

````python
print(num)
````

[함수 안에서 전역 변수 사용하기]
 - global 키워드 사용 

````python
num = 7         
def func():
    global num      # 이 함수 내에서 접근하는 num은 전역변수임을 알려줌
    num = 0         
    print(num)      # 전역변수 num 출력
func()
````

````python
print(num)
````

1. 파일 열기



```
#읽기용
파일변수 = open("파일명","r") 
#쓰기용
파일변수 = open("파일명","w")
```

|파일 모드|설명|
|--|--|
|r|읽기, 해당 경로에 파일이 없으면 에러|
|w|쓰기, 기존 내용 모두 삭제되고 처음부터 기록|
||해당 경로에 파일이 없으면 파일 생성|
|a|쓰기, 현재 파일 내용 뒤에 기록|
||해당 경로에 파일이 없으면 파일 생성|
|r+|읽기 쓰기 겸용|


2. 읽기 및 쓰기 작업

- 쓰기


```
 f = open(“C:\\python\\food.txt”, ‘w’)     # 파일 열기. 경로까지 다 써줘야 함. 그 같은 경로에 새로운 파일을 생성해야 하기 때문
 # 쓰기 형식: 파일변수.write(기록할 문자열)

```



- 읽기



```

 # 읽기 형식: 파일변수.읽기함수()
 data = f.read()
```



|함수|설명|
|--|--|
|read()|파일의 내용 전체를 한꺼번에 읽음|
|readline()|호출할 때마다 파일의 내용을 줄 단위로 읽음|
|readlines()|파일의 모든 내용을 읽어 줄 단위로 리스트 항목을 만들어줌|
